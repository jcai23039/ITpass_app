```python
#51~60

students = [
    ["Alice", 80, 90],
    ["Bob", 60, 55],
    ["Charlie", 95, 100]
]

result_lst = []

for student in students:

    name = student[0]

    total = sum(student[1:])
    average = total / 2

    if average >= 70:
        result = "pass"
    else:
        result = "fail"

    result_dict = {'name': name, 'average': average, 'result': result}
    result_lst.append(result_dict)

for result in result_lst:
    print(f"{result['name']}: {result['average']} {result['result']}")

passed = [result["name"] for result in result_lst if result['result'] == 'pass']
print(f"合格者: {', '.join(passed)}")
    
```

```python
#61~70

def calc_average(*scores):
    return sum(scores) / len(scores)

class Student:
    def __init__(self, name, scores):
         self.name = name
         self.scores = scores

    def average(self):
        return calc_average(*self.scores)
    
    def get_status(self):
        if self.average() >= 70:
            return "pass"
        else:
            return "fail"
        
    def show(self):
        avg = self.average()
        status = self.get_status()
        print(f"{self.name}: {avg} {status}")

student = Student("Alice", [80, 90])
student.show()
```

```python
import csv
import logging
import sys
from pprint import pprint

logging.basicConfig(
    filename="program.log", level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s", encoding="utf-8"
)

csv_file = "students.csv"

input_name = input("Name: ")
input_score = input("Score: ")
student_data = {"name": input_name, "score": input_score}

with open(csv_file, mode="w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "score"])
    writer.writeheader()
    writer.writerow(student_data)

students_list = []
with open(csv_file, mode="r", encoding="utf-8") as f:
    for row in csv.DictReader(f):
        students_list.append(dict(row))

print(f"Name: {input_name}\nScore: {input_score}")
pprint(students_list)
```
