```python
#問題1

students = [
    {"name": "Alice", "scores": [80, 90, 70]},
    {"name": "Bob", "scores": [50, 60, 55]},
    {"name": "Charlie", "scores": [100, 95, 90]},
    {"name": "Dave", "scores": [65, 70, 60]}
]

for student in students:
    scores = student["scores"]
    name = student["name"]
    average = sum(scores) / len(scores)
    
    if average >=70:
        result = "pass"
        passed_students = []
        passed_students.append(name)
    else:
        result = "fail"
    
    print(f"{name}: {average:.1f} {result}")


print(f"合格者:　{','.join(passed_students)}")        
```

```python
#問題2

stock = {
    "apple": 10,
    "banana": 5,
    "orange": 8
}

orders = [
    {"name": "apple", "count": 3},
    {"name": "banana", "count": 6},
    {"name": "orange", "count": 2},
    {"name": "grape", "count": 4}
]

for order in orders:
    name = order["name"]
    count = order["count"]
  
    if name not in stock:
        print(f"{name} は取り扱っていません")
    else:
        name in stock
        if stock[name] >= count:
            stock[name] -= count
            print(f"{name} を {count}個 注文しました")
        else:
            print(f"{name} は在庫が足りません")

       
print(f"残り在庫： {stock}")
```
